### CSV Data Cleaning

```python
#dropping columns with irrelevant data

cleaned_mojo_df= mojo_budget_df.drop(columns= ['movie_id','director', 'writer', 'producer', 'composer', 'cinematographer', 'main_actor_1', 'main_actor_2', 'main_actor_3', 'main_actor_4', 'genre_2', 'genre_3', 'genre_4', 'link' ])

cleaned_mojo_df.head()
```

```python
cleaned_imdb_df= imdb_movie_df.drop(columns= ['Rank', 'Genre', 'Description', 'Director', 'Actors', 'Year', 'Runtime (Minutes)', 'Revenue (Millions)', 'Metascore'])
cleaned_imdb_df.head()
```

```python
# Change column name from title to movie_title
```

```python
cleaned_imdb_df.rename(columns={'Title': 'movie_title'}, inplace=True)
cleaned_imdb_df.head()
```

```python
cleaned_metacritic_df= metacritic_df.drop(columns= ['release_date', 'genre', 'meta_mixed', 'userscore'])
cleaned_metacritic_df.head()
```

```python
# Merge dfs on movie_title
merged_df = pd.merge(cleaned_mojo_df, cleaned_imdb_df, on="movie_title")
merged_df.head()
```
